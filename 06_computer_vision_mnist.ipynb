# သင်ခန်းစာ (၆) - Computer Vision (လက်ရေးဂဏန်းများ ခွဲခြားမည့် AI)

### 🧠 Logic
ဒီနေ့မှာ လူတွေ လက်နဲ့ရေးထားတဲ့ ဂဏန်းပုံလေးတွေကို ကြည့်ပြီး ဘာဂဏန်းလဲ (0, 1, 2, ... 9) ဆိုတာ ခွဲခြားပေးမယ့် AI ကို PyTorch နဲ့ ဆောက်ပါမယ်။
အဆင့်ဆင့် လုပ်ဆောင်ချက်များ -
1. **Torchvision အသုံးပြုခြင်း** - ကမ္ဘာကျော် MNIST ရုပ်ပုံဒေတာအတွဲကို အွန်လိုင်းကနေ အလိုအလျောက် ဒေါင်းလုဒ်ဆွဲခြင်း။
2. **DataLoader ပြင်ဆင်ခြင်း** - ပုံပေါင်း ၆ သောင်းလုံးကို စက်ထဲ တစ်ပြိုင်တည်း မထည့်ဘဲ ပုံ ၆၄ ပုံစီ (Batch Size) ခွဲပြီး စနစ်တကျ ထည့်သွင်းခြင်း။
3. **CNN Model ဆောက်ခြင်း** - `nn.Conv2d` (ရုပ်ပုံဖတ်စနစ်) များကို သုံးပြီး AI ဦးနှောက် ဒီဇိုင်းဆွဲခြင်း။

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms

# ၁။ ရုပ်ပုံဒေတာများကို ဒေါင်းလုဒ်ဆွဲခြင်းနှင့် ပုံမှန်ဒေတာအဖြစ် ပြောင်းလဲခြင်း (Transform)
transform = transforms.Compose([
    transforms.ToTensor(), # ရုပ်ပုံကို PyTorch Tensor ဖြစ်အောင် ပြောင်းမယ်
    transforms.Normalize((0.5,), (0.5,)) # ဒေတာတွေ တွက်ရလွယ်အောင် မျှခြေညှိမယ်
])

# MNIST လက်ရေးဂဏန်း Dataset ကို ဒေါင်းလုဒ်ဆွဲခြင်း
train_dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)

# ၂။ DataLoader ဆောက်ခြင်း (ပုံ ၆၄ ပုံစီ ခွဲပြီး စက်ထဲ ထည့်မယ့်စနစ်)
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=64, shuffle=True)

print("--- အဆင့် (၁) နှင့် (၂) ရုပ်ပုံဒေတာများ အဆင်သင့်ဖြစ်ပါပြီ ---")
print(f"စုစုပေါင်း ရုပ်ပုံအရေအတွက်: {len(train_dataset)} ပုံ")
```

### 🔄 အဆင့် (၃) - CNN (Convolutional Neural Network) Model ဆောက်ခြင်း
ရုပ်ပုံတွေရဲ့ Feature (အစွန်းအနားတွေ၊ ကောက်ကြောင်းတွေ) ကို ထိရောက်စွာ ဖတ်နိုင်ဖို့ `nn.Conv2d` ကို သုံးပါမယ်။ ပြီးရင်တော့ နောက်ဆုံးအဆင့်မှာ ဂဏန်း ၁၀ လုံးထဲက တစ်လုံးလုံး ဖြစ်နိုင်ခြေကို တွက်ထုတ်ဖို့ `nn.Linear` နဲ့ ချိတ်ဆက်ပါမယ်။

```python
# ၃။ CNN Model Class တစ်ခု ဆောက်ခြင်း
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        # Convolution Layer: ရုပ်ပုံထဲက ပုံသဏ္ဍာန်တွေကို ဖတ်တဲ့နေရာ
        self.conv = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, stride=1, padding=1)
        self.relu = nn.ReLU()
        # MaxPool: ပုံကို သေးချလိုက်ပြီး အဓိကအချက်တွေကိုပဲ ချုံ့ယူတဲ့နေရာ
        self.pool = nn.Pool2d = nn.MaxPool2d(kernel_size=2, stride=2) 
        
        # Fully Connected Layer: နောက်ဆုံး ဂဏန်း ၁၀ မျိုးထဲက ရွေးချယ်မယ့်နေရာ
        self.fc = nn.Linear(16 * 14 * 14, 10) 

    def forward(self, x):
        x = self.conv(x)
        x = self.relu(x)
        x = self.pool(x)
        x = x.view(x.size(0), -1) # Matrix ကို တစ်တန်းတည်းဖြစ်အောင် ဖြန့်ချလိုက်တာ (Flatten)
        x = self.fc(x)
        return x

model = SimpleCNN()
print("--- အဆင့် (၃) CNN Model ဆောက်လုပ်မှု အောင်မြင်ပါသည် ---")
print(model)
```

### 🎯 အဆင့် (၄) - Loss Function နှင့် Training Loop
ဒီတစ်ခါမှာတော့ အမျိုးအစား ခွဲခြားတာ (Classification) ဖြစ်လို့ Loss Function အဖြစ် **`nn.CrossEntropyLoss()`** ကို သုံးရပါမယ်။ ဒေတာပုံစံ ပေါင်း ၆ သောင်းရှိလို့ အချိန်အများကြီး မပေးရအောင် လောလောဆယ် `1 Epoch` (ဒေတာအားလုံးကို တစ်ခေါက်ပဲ လှည့်ပတ်သင်ကြားခြင်း) ပဲ လုပ်ပါမယ်။

```python
# ၄။ Loss, Optimizer သတ်မှတ်ခြင်း နှင့် Training စတင်ခြင်း
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001) # SGD ထက် ပိုမြန်တဲ့ Adam Optimizer ကို သုံးထားပါတယ်

print("--- 🏋️ AI ရုပ်ပုံများကို စတင်ကြည့်ရှု လေ့ကျင့်နေပါပြီ (ခဏစောင့်ပေးပါ) ---")

for batch_idx, (images, labels) in enumerate(train_loader):
    # Forward Pass
    outputs = model(images)
    loss = criterion(outputs, labels)
    
    # Backward and optimize
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # ပုံပေါင်း ၁ သောင်းခွဲခန့် (Batch ပေါင်း ၂၀၀၀ စီ) ပြည့်တိုင်း အခြေအနေ ထုတ်ကြည့်မယ်
    if (batch_idx + 1) % 200 == 0:
        print(f"Batch [{batch_idx + 1}/{len(train_loader)}], Loss: {loss.item():.4f}")

print("--- 🏁 ၁ Epoch လေ့ကျင့်မှု ပြီးဆုံးပါပြီ။ AI တော်တော်လေး နားလည်သွားပါပြီ ---")
```
