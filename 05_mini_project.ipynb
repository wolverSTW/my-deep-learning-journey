# သင်ခန်းစာ (၅) - နမူနာ ပရောဂျက် (အိမ်ဈေးနှုန်း ခန့်မှန်းပေးမယ့် AI)

### 🧠 Logic
ဒီ ပရောဂျက်မှာ အိမ်အကျယ်အဝန်း (Sqft) ကိုကြည့်ပြီး အိမ်ဈေးနှုန်း (သိန်းပေါင်း) ဘယ်လောက်ရှိလဲ ခန့်မှန်းပေးမယ့် AI ကို PyTorch နဲ့ တည်ဆောက်ပါမယ်။
အဆင့် ၄ ဆင့်ဖြင့် သွားပါမယ် -
1. **Data ပြင်ဆင်ခြင်း** - နမူနာ ဒေတာအတွဲများ တည်ဆောက်ခြင်း။
2. **Model ဆောက်ခြင်း** - `nn.Linear` သုံးပြီး AI ဦးနှောက်ဆောက်ခြင်း။
3. **Loss & Optimizer သတ်မှတ်ခြင်း** - MSE Loss နဲ့ SGD (Gradient Descent) ကို သုံးခြင်း။
4. **Training Loop** - AI ကို အခါပေါင်း ၁၀၀ တိတိ အမှားပြင်ခိုင်းပြီး သင်ကြားပေးခြင်း (Training)။

```python
import torch
import torch.nn as nn
import torch.optim as optim

# ၁။ Data ပြင်ဆင်ခြင်း (အိမ်အကျယ်အဝန်း X နှင့် တကယ့်အိမ်ဈေးနှုန်း Y)
# [1000 sqft, 1500 sqft, 2000 sqft, 2500 sqft]
X_train = torch.tensor([[1000.0], [1500.0], [2000.0], [2500.0]], dtype=torch.float32)
# [သိန်း ၅၀၀, သိန်း ၇၅၀၀, သိန်း ၁၀၀၀, သိန်း ၁၂၅၀] 
Y_train = torch.tensor([[500.0], [750.0], [1000.0], [1250.0]], dtype=torch.float32)

print("--- အဆင့် (၁) ပရောဂျက်အတွက် ဒေတာများ အဆင်သင့်ဖြစ်ပါပြီ ---")
print(f"X_train (Inputs) Shape: {X_train.shape}")
print(f"Y_train (Targets) Shape: {Y_train.shape}")
```

### 🔄 အဆင့် (၂) နှင့် (၃) - Model ဆောက်ခြင်းနှင့် Tool များ ပြင်ဆင်ခြင်း
* Input က အိမ်အကျယ်အဝန်း (၁) ခုဖြစ်ပြီး Output က အိမ်ဈေးနှုန်း (၁) ခု ဖြစ်လို့ `nn.Linear(1, 1)` ကို သုံးပါမယ်။
* အမှားတိုင်းတာဖို့ **`nn.MSELoss()`** ကို သုံးပြီး၊ Weights တွေကို အလိုအလျောက် ညှိပေးဖို့ **`optim.SGD`** (Stochastic Gradient Descent) ကို သုံးပါမယ်။ Learning Rate ကို `0.0000001` လို့ သေးသေးလေး ပေးထားပါမယ် (ဒေတာတွေ ဂဏန်းကြီးလို့ပါ)။

```python
# ၂။ Model ဆောက်ခြင်း
model = nn.Linear(in_features=1, out_features=1)

# ၃။ Loss Function နှင့် Optimizer (Weights ညှိပေးမယ့်ကိရိယာ) သတ်မှတ်ခြင်း
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=1e-7)

print("--- အဆင့် (၂) နှင့် (၃) Model နှင့် သင်ကြားရေး Tools များ ပြင်ဆင်ပြီး ---")
```

### 🎯 အဆင့် (၄) - Training Loop (AI ကို အခါ ၁၀၀ တိတိ လေ့ကျင့်ပေးခြင်း)
အဆင့်တွေကတော့ -
1. `model(X_train)`: AI ကို ခန့်မှန်းခိုင်းမယ် (Forward Pass)
2. `criterion(pred, true)`: အမှားတွက်မယ် (Loss Calculation)
3. `loss.backward()`: ဘယ်နေရာမှာ ဘယ်လောက်မှားလဲ နောက်ကြောင်းပြန်တွက်မယ် (Backpropagation)
4. `optimizer.step()`: Weights တွေကို လိုက်ညှိမယ် (Gradient Descent)

```python
# ၄။ AI ကို သင်ကြားပေးခြင်း (Training) - အခါ ၁၀၀ တိတိ လှည့်ပတ်သင်ပါမယ်
epochs = 100

print("--- 🏋️ AI စတင်လေ့ကျင့်နေပါပြီ ---")
for epoch in range(epochs):
    # (က) AI ကို အဖြေခန့်မှန်းခိုင်းခြင်း
    predictions = model(X_train)
    
    # (ခ) အမှား (Loss) တွက်ချက်ခြင်း
    loss = criterion(predictions, Y_train)
    
    # (ဂ) အမှားဟောင်းတွေကို ရှင်းထုတ်ပြီး နောက်ကြောင်းပြန်တွက်ခြင်း
    optimizer.zero_grad()
    loss.backward()
    
    # (ဃ) Weights များကို ညှိလိုက်ခြင်း
    optimizer.step()
    
    # အကြိမ် ၂၀ ပြည့်တိုင်း အမှားကျမကျ ထုတ်ကြည့်ခြင်း
    if (epoch + 1) % 20 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss (အမှားပမာဏ): {loss.item():.4f}")
```

### 🏁 AI တကယ် တော်၊ မတော် စမ်းသပ်ခြင်း (Testing)
အခုဆိုရင် AI က အခါ ၁၀၀ လေ့ကျင့်ပြီးပါပြီ။ မော်ဒယ် မမြင်ဖူးသေးတဲ့ **၁၈၀၀ စတုရန်းပေ (1800 sqft)** ရှိတဲ့ အိမ်တစ်အိမ်ကို ပေးပြီး ဈေးနှုန်း ခန့်မှန်းခိုင်းကြည့်ပါမယ်။ (တကယ်ထွက်ရမယ့် မှန်ကန်တဲ့အဖြေက သိန်း ၉၀၀ ဝန်းကျင် ဖြစ်ရပါမယ်)။

```python
# စမ်းသပ်မည့် ဒေတာအသစ် [1800 sqft]
new_house = torch.tensor([[1800.0]], dtype=torch.float32)

# AI ကို ခန့်မှန်းခိုင်းခြင်း (Evaluation Mode)
model.eval()
with torch.no_grad(): # စမ်းသပ်ချိန်မို့ Weight တွေ ထပ်မညှိခိုင်းတော့ပါဘူး
    ai_answer = model(new_house)

print("--- 🏁 AI ၏ နောက်ဆုံး ခန့်မှန်းချက် ရလဒ် ---")
print(f"အိမ်အကျယ်အဝန်း: 1800 sqft")
print(f"AI ခန့်မှန်းပေးသော အိမ်ဈေးနှုန်း: သိန်း {ai_answer.item():.2f}")
```
