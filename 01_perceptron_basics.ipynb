# သင်ခန်းစာ (၁) - Perceptron (ပါဆက်ပထရွန်) အခြေခံ

### 🧠 Logic ကမ္ဘာ
Perceptron ဆိုတာ Deep Learning ရဲ့ အခြေခံအကျဆုံး အစိတ်အပိုင်း (ဦးနှောက်ဆဲလ်တစ်ခုတည်း) ပါပဲ။
သူ့မှာ အဓိက အပိုင်း ၃ ပိုင်း ရှိပါတယ်။
1. **Inputs ($x$)**: ဝင်လာတဲ့ အချက်အလက်တွေ (Data)
2. **Weights ($w$) & Bias ($b$)**: ဘယ်အချက်အလက်က ပိုအရေးကြီးလဲဆိုတဲ့ 'အလေးချိန်' နဲ့ မျှခြေညှိပေးတဲ့ 'ကိန်းသေ' (AI က ဒါတွေကိုပဲ လိုက်ညှိပြီး သင်ယူတာပါ)
3. **Activation Function**: ရလာတဲ့ ပေါင်းလဒ်ကို $1$ သို့မဟုတ် $0$ (Yes/No) ဆုံးဖြတ်ချက် ဖြစ်အောင် ပြောင်းပေးတဲ့ ဇကာကွက်

```python
# ၁။ အချက်အလက်များ (Inputs) သတ်မှတ်ခြင်း
# ဥပမာ- မိုးမရွာဘူး (0)၊ သူငယ်ချင်းလာမယ် (1)၊ ပိုက်ဆံရှိတယ် (1)
inputs = [0, 1, 1]

# ၂။ အလေးချိန် (Weights) နှင့် Bias သတ်မှတ်ခြင်း
weights = [0.7, 0.2, 0.9]
bias = -1.1  # တွက်ချက်မှုရလဒ် -0.1 (သုညအောက်ငယ်) ဖြစ်အောင် ညှိထားတာပါ

print("--- အဆင့် (၁) နှင့် (၂) အချက်အလက်များ ပြင်ဆင်ပြီး ---")
print(f"Inputs: {inputs}")
print(f"Weights: {weights}")
print(f"Bias: {bias}")
```

### 🔄 အဆင့် (၃) - Weighted Sum တွက်ချက်ခြင်း
Formula ကတော့ $\text{Sum} = (x_1 \cdot w_1) + (x_2 \cdot w_2) + (x_3 \cdot w_3) + b$ ဖြစ်ပါတယ်။
ကွန်ပျူတာက Input တစ်ခုချင်းစီကို သူ့ Weight နဲ့မြှောက်ပြီး အကုန်လိုက်ပေါင်းပါမယ်။ ပြီးရင် Bias ထည့်ပေါင်းပါမယ်။

```python
# ၃။ Weighted Sum တွက်ခြင်း
weighted_sum = 0

# Input နဲ့ Weight ကို တစ်ခုချင်းစီ မြှောက်ပြီး ပေါင်းမယ်
for i in range(len(inputs)):
    weighted_sum += inputs[i] * weights[i]

# နောက်ဆုံးမှ Bias ကို ပေါင်းထည့်မယ်
weighted_sum += bias

print(f"တွက်ချက်မှုရလဒ် (Weighted Sum): {weighted_sum:.1f}")
```

### 🎯 အဆင့် (၄) - Activation Function (Step Function)
ရလာတဲ့ `weighted_sum` က $0$ ထက်ကြီးရင် သို့မဟုတ် ညီရင် $1$ (ဘောလုံးသွားကန်မယ်)၊ $0$ ထက်ငယ်ရင် $0$ (အိမ်မှာပဲနေမယ်) လို့ ဆုံးဖြတ်ချက် ချပါမယ်။

```python
# ၄။ Activation Function ဇကာကွက်ထဲ ဖြတ်ခြင်း
if weighted_sum >= 0:
    output = 1  # Yes
else:
    output = 0  # No

print(f"Activation Function မှ ထွက်လာသော ရလဒ် (Output): {output}")
if output == 1:
    print("ဆုံးဖြတ်ချက်: ဘောလုံးသွားကန်ပါမယ်။")
else:
    print("ဆုံးဖြတ်ချက်: အိမ်မှာပဲ နေပါတော့မယ်။")
```
