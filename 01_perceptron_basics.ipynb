```python
# ၁။ အချက်အလက်များ (Inputs) သတ်မှတ်ခြင်း
# မိုးမရွာဘူး (0)၊ သူငယ်ချင်းလာမယ် (1)၊ ပိုက်ဆံရှိတယ် (1)
inputs = [0, 1, 1]

# ၂။ အလေးချိန် (Weights) နှင့် Bias သတ်မှတ်ခြင်း
# တွက်ချက်မှုရလဒ် -0.1 ဖြစ်အောင် Bias ကို -1.1 လို့ ပြောင်းပေးလိုက်ပါတယ်
weights = [0.7, 0.2, 0.9]
bias = -1.1

# ၃။ Weighted Sum တွက်ခြင်း (Inputs * Weights + Bias)
weighted_sum = 0
for i in range(len(inputs)):
    weighted_sum += inputs[i] * weights[i]

weighted_sum += bias  # Bias ပေါင်းထည့်ခြင်း

# ၄။ Activation Function (Step Function ဇကာကွက်)
if weighted_sum >= 0:
    output = 1  # Yes (ဘောလုံးသွားကန်မယ်)
else:
    output = 0  # No (အိမ်မှာပဲနေမယ်)

# ရလဒ် ထုတ်ကြည့်ခြင်း
print(f"တွက်ချက်မှုရလဒ် (Weighted Sum): {weighted_sum:.1f}")
print(f"Activation Function မှ ထွက်လာသော ရလဒ် (Output): {output}")
```
