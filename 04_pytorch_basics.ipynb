# သင်ခန်းစာ (၄) - PyTorch Framework အခြေခံ

### 🧠 Logic
လုပ်ငန်းခွင်သုံး **PyTorch** ကို စတင်လေ့လာပါမယ်။
ဒီကုဒ်မှာ-
* PyTorch ရဲ့ အခြေခံ `Tensor` တည်ဆောက်ပုံကို လေ့လာမယ်။
* ရှေ့မှာ သင်ခဲ့တဲ့ Linear Layer (Inputs * Weights + Bias) ကို PyTorch ရဲ့ `nn.Linear` သုံးပြီး တစ်ကြောင်းတည်းနဲ့ ဆောက်ပါမယ်။

```python
import torch
import torch.nn as nn  # nn ဆိုတာ Neural Network ရဲ့ အတိုကောက်ပါ

# ၁။ PyTorch Tensor ဆောက်ခြင်း (NumPy array နဲ့ တူတူပါပဲ)
# အိမ်တစ်အိမ်ရဲ့ အချက်အလက် (ဥပမာ- [အိပ်ခန်းအရေအတွက်, အကျယ်အဝန်း])
house_features = torch.tensor([[3.0, 1500.0]])

print("--- အဆင့် (၁) PyTorch Tensor ပြင်ဆင်ပြီး ---")
print(f"Input Tensor: {house_features}")
print(f"Data Type: {house_features.dtype}")
print(f"Tensor Shape: {house_features.shape}")
```

### 🔄 အဆင့် (၂) - PyTorch ဖြင့် Layer တည်ဆောက်ခြင်း
ရှေ့သင်ခန်းစာ ၂ တုန်းကလို `weights` တွေ `bias` တွေ ကိုယ်တိုင်လိုက်ရေးပြီး `np.dot` လုပ်နေစရာ မလိုတော့ပါဘူး။
`nn.Linear(in_features, out_features)` ကို သုံးပြီး အလိုအလျောက် တွက်ခိုင်းပါမယ်။
* `in_features=2` (ဝင်လာမယ့် Input အချက်အလက် ၂ ခု)
* `out_features=1` (ထွက်လာမယ့် အဖြေ ၁ ခု - ဥပမာ အိမ်ဈေးနှုန်း)

```python
# ၂။ Linear Layer တစ်ခု တည်ဆောက်ခြင်း
# PyTorch က နောက်ကွယ်မှာ Weights နဲ့ Bias တွေကို သူ့အလိုလို Random တန်ဖိုးတွေနဲ့ ဆောက်ပေးသွားပါတယ်
linear_layer = nn.Linear(in_features=2, out_features=1)

print("--- အဆင့် (၂) PyTorch ၏ Linear Layer အတွင်းပိုင်း ---")
print(f"အလိုအလျောက် ဆောက်ပေးထားသော Weights:\n {linear_layer.weight}")
print(f"အလိုအလျောက် ဆောက်ပေးထားသော Bias: {linear_layer.bias}")
```

### 🎯 အဆင့် (၃) - Forward Pass (ရှေ့သို့ တွန်းလွှတ်တွက်ချက်ခြင်း)
Input Tensor ကို ဆောက်ထားတဲ့ Linear Layer ထဲ ထည့်လိုက်ရုံနဲ့ $x \cdot w + b$ ကို အနောက်ကွယ်မှာ တွက်ချက်ပြီး အဖြေထုတ်ပေးသွားမှာ ဖြစ်ပါတယ်။

```python
# ၃။ Layer ထဲသို့ Input ထည့်၍ အဖြေထုတ်ခြင်း
prediction = linear_layer(house_features)

print("--- အဆင့် (၃) 🏁 မော်ဒယ်၏ ခန့်မှန်းချက်ရလဒ် ---")
print(f"Predicted House Price: {prediction.item():.2f}") 
# .item() က PyTorch Tensor ထဲက ဂဏန်းသီးသန့်ကို ဆွဲထုတ်ပေးတာပါ
```
