# သင်ခန်းစာ (၂) - Multi-Layer Neural Network အခြေခံ

### 🧠 Logic (Real World)
အာရုံကြောလေးတွေ အများကြီး စုပေါင်းပြီး အလွှာ (Layers) လိုက် အလုပ်လုပ်ပုံကို လေ့လာပါမယ်။
ဒီကုဒ်မှာ-
* Input အချက်အလက် (၂) ခု ရှိမယ်။
* Hidden Layer ထဲမှာ အာရုံကြော (Neuron) (၃) ခု ရှိမယ်။
* Output Layer မှာ ဆုံးဖြတ်ချက် (၁) ခု ထွက်မယ်။

Matrix တွက်ချက်မှုတွေ ပါလာမှာဖြစ်လို့ Python ရဲ့ **NumPy** Library ကို စတင်အသုံးပြုပါမယ်။

```python
import numpy as np # သင်္ချာ Matrix တွက်ချက်မှုအတွက် မရှိမဖြစ် Library ဖြစ်ပါတယ်။

# ၁။ Inputs ပြင်ဆင်ခြင်း (ဥပမာ- အချက်အလက် ၂ ခု)
# ဥပမာ- လူတစ်ယောက်ရဲ့ အရပ်နဲ့ အလေးချိန် [အရပ်, အလေးချိန်]
inputs = np.array([1.7, 65.0])

print("--- အဆင့် (၁) Input Data ပြင်ဆင်ပြီး ---")
print(f"Inputs: {inputs} (Shape: {inputs.shape})")
```

### 🔄 အဆင့် (၂) - Hidden Layer (neuron ၃ ခု၏ Weights နှင့် Biases)
neurons ၃ ခု ရှိတဲ့အတွက် တစ်ခုချင်းစီမှာ Input (၂) ခုအတွက် Weights (၂) ခုစီ ရှိရပါမယ်။ 
ဒါကြောင့် Weights တွေကို Matrix ပုံစံ $[3 \times 2]$ နဲ့ သိမ်းပြီး၊ Bias ကိုတော့ neurons ၃ ခုအတွက် (၃) ခု သီးသန့်ပေးပါမယ်။

```python
# ၂။ Hidden Layer အတွက် Weights နှင့် Biases သတ်မှတ်ခြင်း (နမူနာ ကိန်းဂဏန်းများ)
hidden_weights = np.array([
    [0.1, 0.5],   # Neuron 1 ရဲ့ Weights
    [-0.4, 0.2],  # Neuron 2 ရဲ့ Weights
    [0.6, -0.1]   # Neuron 3 ရဲ့ Weights
])
hidden_biases = np.array([0.1, -0.2, 0.3]) # Neuron ၃ ခုအတွက် Bias ၃ ခု

# Matrix မြှောက်ခြင်း (Dot Product) နှင့် Bias ပေါင်းခြင်းကို NumPy က တစ်ကြောင်းတည်းနဲ့ တွက်ပေးနိုင်ပါတယ်
# Formula: (Inputs * Weights) + Bias
hidden_layer_input = np.dot(hidden_weights, inputs) + hidden_biases


print(hidden_weights)

print("--- Hidden Layer သို့ ဝင်လာသော ပေါင်းလဒ်များ ---")
print(f"Calculated Sums for 3 Neurons: {hidden_layer_input}")
```

### 🎯 အဆင့် (၃) - Activation Function ဖြတ်ခြင်း (ReLU)
ဒီတစ်ခါမှာတော့ Deep Learning မှာ အသုံးအများဆုံး **ReLU (Rectified Linear Unit)** ကို သုံးပါမယ်။
**ReLU ရဲ့ Logic:** သုညထက် ငယ်ရင် သုညဖြစ်သွားမယ်၊ သုညထက်ကြီးရင် အဲဒီအတိုင်း အပြင်ကို ပေးထွက်မယ်။ ($f(x) = \max(0, x)$)

```python
# ReLU Activation Function ကို ဆောက်ခြင်း
def relu(x):
    return np.maximum(0, x)

# Hidden Layer ရဲ့ ရလဒ်ကို ReLU ထဲ ဖြတ်မယ်
hidden_layer_output = relu(hidden_layer_input)

print("--- ReLU ဇကာကွက် ဖြတ်ပြီးနောက် Hidden Layer ရလဒ် ---")
print(f"Hidden Layer Outputs: {hidden_layer_output}")
```

### 🏁 အဆင့် (၄) - Output Layer (နောက်ဆုံး ဆုံးဖြတ်ချက်တွက်ခြင်း)
Hidden Layer က ထွက်လာတဲ့ ရလဒ် (၃) ခုက Output Layer အတွက် Input ပြန်ဖြစ်သွားပါတယ်။
အဲဒီ (၃) ခုကို Weight (၃) ခုနဲ့ မြှောက်၊ Bias ပေါင်းပြီး နောက်ဆုံး Output ထုတ်ပါမယ်။

```python
# Output Layer အတွက် Weights (၃ ခု) နှင့် Bias (၁ ခု)
output_weights = np.array([0.3, -0.5, 0.2])
output_bias = 0.1

# နောက်ဆုံး တွက်ချက်မှု
final_sum = np.dot(output_weights, hidden_layer_output) + output_bias
final_output = relu(final_sum)  # နောက်ဆုံးအဆင့်ကိုပါ ReLU ဖြတ်လိုက်ပါတယ်

print("--- 🏁 နောက်ဆုံးရလဒ် အဖြေ ---")
print(f"Final AI Prediction Output: {final_output:.4f}")
```
