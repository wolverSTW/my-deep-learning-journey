# သင်ခန်းစာ (၃) - Loss Function နှင့် Gradient Descent အခြေခံ

### 🧠 Logic 
AI က အဖြေမှန်ကို မသိမချင်း သူ့ရဲ့ အမှား (Loss) ကို တွက်ပြီး Weights တွေကို ဖြည်းဖြည်းချင်း လိုက်ညှိရပါတယ်။
ဒီကုဒ်မှာ-
* AI ထုတ်ပေးလိုက်တဲ့ အဖြေ (Prediction) ရှိမယ်။
* တကယ့် အဖြေမှန် (Target / Real Value) ရှိမယ်။
* သူတို့ကြားက အမှားကို **Mean Squared Error (MSE)** နည်းနဲ့ တွက်မယ်။
* အမှားနည်းသွားအောင် Weights တွေကို ဖြည်းဖြည်းချင်း ညှိတဲ့ **Gradient Descent** Logic ကို ကုဒ်နဲ့ စမ်းသပ်ပါမယ်။

```python
import numpy as np

# ၁။ AI ရဲ့ ခန့်မှန်းချက် (Prediction) နှင့် တကယ့် အဖြေမှန် (Actual)
# ဥပမာ- အိမ်ဖိုးကို AI က သိန်း ၅၀၀ လို့ ခန့်မှန်းပေမယ့် အဖြေမှန်က သိန်း ၆၀၀ ဖြစ်နေတယ် ဆိုပါစို့
y_pred = np.array([500.0])
y_true = np.array([600.0])

print("--- အဆင့် (၁) ခန့်မှန်းချက်နှင့် အဖြေမှန် ပြင်ဆင်ပြီး ---")
print(f"AI Prediction: {y_pred}")
print(f"Actual Target: {y_true}")
```

### 🔄 အဆင့် (၂) - Loss (အမှား) တွက်ခြင်း
ဒီနေရာမှာ **MSE (Mean Squared Error)** ကို သုံးပါမယ်။ 
Formula ကတော့ $\text{Loss} = (\text{Prediction} - \text{Actual})^2$ ဖြစ်ပါတယ်။
ကွာခြားချက်ကို နှစ်ထပ် (Square) တွက်ရတာက အနှုတ်လက္ခဏာတွေ ပျောက်သွားအောင်နဲ့ အမှားကြီးရင် ကြီးသလို ပြစ်ဒဏ်ကြီးကြီး ပေးနိုင်အောင် ဖြစ်ပါတယ်။

```python
# ၂။ Loss Function ကို တွက်ချက်ခြင်း
def calculate_mse(pred, true):
    return np.mean((pred - true) ** 2)

current_loss = calculate_mse(y_pred, y_true)

print("--- အဆင့် (၂) လက်ရှိ AI ၏ အမှားပမာဏ (Loss) ---")
print(f"Current Loss: {current_loss} (ဒီအမှားကြီးကို သုညဖြစ်အောင် လျှော့ချရပါမယ်)")
```

### 🎯 အဆင့် (၃) - Gradient Descent Logic (Weights များကို ညှိခြင်း)
အမှားနည်းသွားအောင် သင်္ချာနည်းအရ Derivative (Slope) ရှာရပါတယ်။ Slope က Negative ဖြစ်နေရင် Weight ကို တိုးပေးရပြီး၊ Positive ဖြစ်နေရင် လျှော့ပေးရပါတယ်။
ဒီနေရာမှာ **Learning Rate ($\alpha$)** ဆိုတဲ့ ကိန်းသေတစ်ခုကို သုံးပြီး Weights တွေကို တဒုန်းဒုန်းမပြောင်းဘဲ ဖြည်းဖြည်းချင်း (Step အလိုက်) ညှိပေးရပါတယ်။

```python
# နမူနာ Weight တစ်ခု ရှိတယ်ဆိုပါစို့
current_weight = 0.5
learning_rate = 0.01  # AI က ခြေလှမ်း ဘယ်လောက်ကျကျ လှမ်းမလဲဆိုတာ သတ်မှတ်တာပါ

# သင်္ချာ Derivative ရှာလိုက်လို့ ထွက်လာတဲ့ လမ်းညွှန်ချက် (Gradient) 
# (ဒီနေရာမှာ ရိုးရှင်းအောင် ပုံသေ ဥပမာအနေနဲ့ ပေးထားတာပါ)
gradient = -20.0  # Gradient က မိုင်နပ်စ်ဖြစ်နေရင် Weight ကို တိုးပေးရမယ်လို့ ဆိုလိုတယ်

# Weight ကို Update လုပ်ခြင်း Formula: Weight = Weight - (Learning Rate * Gradient)
new_weight = current_weight - (learning_rate * gradient)

print("--- အဆင့် (၃) Gradient Descent ဖြင့် Weight ကို ညှိခြင်း ---")
print(f"Old Weight: {current_weight}")
print(f"New Updated Weight: {new_weight} (Weight တက်လာတဲ့အတွက် နောက်တစ်ခေါက်ဆို အမှားနည်းသွားပါလိမ့်မယ်)")
```
